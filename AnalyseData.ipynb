```python
import pandas as pd
from bokeh.plotting import figure, output_file, show, output_notebook
output_notebook()
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('Sample - Superstore.xls')
```

```python
data.head()
```

```python
data.describe()
```

```python
data.describe(include=object)
```

```python
data.info()
```

```python
data.set_index('Order Date', inplace=True)
```

```python
data.head()
```

```python
y = data.Sales.resample('MS').count()
```

```python
p = figure(plot_width=800, plot_height= 400, x_axis_type='datetime')
p.line(y.index, y.values)
show(p)
```

```python
region_count = data.groupby('Region').count()
region_count.index
p = figure(plot_width = 800, plot_height = 400, x_range=list(region_count.index))
p.vbar(x=region_count.index, top=region_count.values[:,0], width=0.9)

show(p)
```

```python
Segment_count = data.groupby('Segment').count()
p = figure(plot_width = 800, plot_height = 400, x_range=list(Segment_count.index))
p.vbar(x=Segment_count.index, top=Segment_count.values[:,0], width=0.9)
show(p)
```

```python

shipMode_count = data.groupby('Ship Mode').count()
p = figure(plot_width = 800, plot_height = 400, x_range=list(shipMode_count.index))
p.vbar(x=shipMode_count.index, top=shipMode_count.values[:,0], width=0.9)
show(p)
```

```python
customer_shop_count = data.groupby('Customer ID').count()
```

```python
customer_shop_counts = data['Customer ID'].value_counts()
```

```python
p= figure(plot_width=400, plot_height=400)
p.quad(source = customer_shop_counts.values,bottom=0, 
           top='count', 
           left='left', 
           right='right')
show(p)
```
